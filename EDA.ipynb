```python
import openpyxl

df_country = pd.read_excel('Dataset/Country-Code.xlsx', engine='openpyxl')
df_country.head()
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
final_df.head()
```

```python
final_df.shape
```

```python
final_df.columns
```

```python
final_df['Country'].value_counts()
```

```python
country_names = final_df['Country'].value_counts().index
country_names
```

```python
country_val = final_df['Country'].value_counts().values
country_val
```

```python
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.1f%%')
plt.show()
```

OBSERVATIONS:
- Zomato's maximum orders and transactions are from India.
- After that USA then United  Kningdom.

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text',]).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

OBSERVATIONS:
- Rating between 1.8 - 2.4 are Poor
- Rating between 2.5 - 3.4 are Average
- Rating between 3.5 - 3.9 are Good
- Rating between 4.0 - 4.4 are Very Good
- Rating between 4.5 - 4.9 are Excellent

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings, hue='Rating color')
plt.show()
```

OBSERVATIONS:
- Not rated count is very high (more than 2000)
- Maximum rating is good (2.5 to 3.4)

## Conclusion & Key Takeaways

This exploratory data analysis of the Zomato dataset has revealed several important insights about the global food delivery and restaurant landscape:

###  Geographic Insights
- **India emerges as the dominant market** with the highest number of orders and transactions, significantly outpacing other countries
- The USA and United Kingdom follow as secondary markets, indicating strong presence in Western economies
- This geographic concentration suggests Zomato's strategic focus on the Indian subcontinent while maintaining presence in key international markets

###  Rating Patterns
- The majority of restaurants fall in the **"Good" rating category (2.5-3.4)**, suggesting room for quality improvement
- A concerning **2000+ restaurants remain unrated**, representing a significant data gap that could impact decision-making
- The rating distribution shows a normal curve with fewer restaurants at the extreme ends (Poor and Excellent)
- India has the maximum number of zero-rated restaurants, while the UK has the minimum, possibly indicating different user behavior or data collection practices

###  Online Delivery Landscape
- **Only 2 countries (India & UAE) offer online delivery**, highlighting limited digital service penetration
- India dominates online delivery with **2000+ delivery regions** compared to UAE's 25+ regions
- This presents a significant **expansion opportunity** for Zomato in other markets like the USA and UK

###  Cuisine Preferences
- Analysis of top 10 cuisines reveals customer preferences and can guide restaurant partners on menu optimization
- Diverse cuisine options indicate a multicultural customer base with varied tastes

###  Business Intelligence
- Currency mapping across countries provides valuable context for pricing strategies and market positioning
- Understanding regional payment preferences can optimize transaction processes

###  Data Quality Observations
- **Cuisines column has 9 missing values** - relatively clean dataset overall
- High number of unrated restaurants suggests need for better user engagement strategies
- Data completeness varies by country, indicating different operational maturity levels
